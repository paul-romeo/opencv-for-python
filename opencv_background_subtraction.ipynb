```python
#################################################################################
# Process the background mask using the cv2.bgsegm.createBackgroundSubtractorMOG()
#  Note: There is a need of "bgsegm" attribute for this method 
#################################################################################
import cv2 
import numpy as np 

cap = cv2.VideoCapture("vtest.avi")

# Create the background segment 
fgbg = cv2.bgsegm.createBackgroundSubtractorMOG()

while True: 
    ret, frame = cap.read()
    if frame is None: 
        break 

    
    # Process the foreground mask 
    fgmask = fgbg.apply(frame)

    # Display the original frame and the foreground mask 
    cv2.imshow("Original Frame", frame)
    cv2.imshow("Foreground Mask", fgmask)
    

    keyboard = cv2.waitKey(30)
    if keyboard == ord('q') or keyboard == 27: 
        break  # exit the while loop 

cap.release()
cv2.destroyAllWindows()

```

```python
##########################################################################################
# Process the background mask using the cv2.createBackgroundSubtractorMOG2() method
#  Note: there is no "segm" attribute in the method 
##########################################################################################
import cv2 
import numpy as np 

cap = cv2.VideoCapture("vtest.avi")

# Create the background segment 
fgbg = cv2.createBackgroundSubtractorMOG2(detectShadows=True)

while True: 
    ret, frame = cap.read()
    if frame is None: 
        break 

    
    # Process the foreground mask 
    fgmask = fgbg.apply(frame)

    # Display the original frame and the foreground mask 
    cv2.imshow("Original Frame", frame)
    cv2.imshow("Foreground Mask", fgmask)
    

    keyboard = cv2.waitKey(30)
    # Press 'q' to exit the app 
    if keyboard == ord('q') or keyboard == 27: 
        break   

cap.release()
cv2.destroyAllWindows()
```

```python
##########################################################################################
# Process the background mask using the cv2.bgsegm.createBackgroundSubtractorGMG() method
#  Note: there is a need of "bgsegm" attribute for this method 
##########################################################################################
#  NOT WORKING ON THIS METHOD even the kernel was set to the fgmask 
import cv2 
import numpy as np 

cap = cv2.VideoCapture("vtest.avi")

# Create the kernel 
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))

# Create the background segment 
fgbg = cv2.bgsegm.createBackgroundSubtractorGMG()

while True: 
    ret, frame = cap.read()
    if frame is None: 
        break 

    
    # Process the foreground mask 
    fgmask = fgbg.apply(frame)
    fgmask = cv2.morphologyEx(fgmask, cv2.MORPH_OPEN, kernel)

    # Display the original frame and the foreground mask 
    cv2.imshow("Original Frame", frame)
    cv2.imshow("Foreground Mask", fgmask)
    

    keyboard = cv2.waitKey(30)
    # Press 'q' to exit the app 
    if keyboard == ord('q') or keyboard == 27: 
        break   

cap.release()
cv2.destroyAllWindows()
```
