```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 

# Open the image and convert color from BGR to RGB 
img = cv2.imread("opencv-logo.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5), np.float32) / 25

convolution = cv2.filter2D(img, -1, kernel)
blur = cv2.blur(img, (5,5))

# Display the images 
titles = ["Image", "2-D Convolution", "Blur"]
images = [img, convolution, blur]

for i in range(3):
    plt.subplot(1,3,i+1), plt.imshow(images[i], "gray")
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()

```

```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 

# Open the image and convert color from BGR to RGB 
img = cv2.imread("lena.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5), np.float32) / 25

convolution = cv2.filter2D(img, -1, kernel)
blur = cv2.blur(img, (5,5))

# Gaussian blur filter tends to be better than the blur filter 
gaussian_blur = cv2.GaussianBlur(img, (5,5), 0)

# Median filter removes salt & pepper noise 
median = cv2.medianBlur(img, 5)

# Bilateral filter (high pass filter) sharpens (the edge) of an image 
bilateral = cv2.bilateralFilter(img,9,75,75)


# Display the images 
titles = ["Image", "2-D Convolution", "Blur", "Gaussian Blur", "Median", "Bilateral Filter"]
images = [img, convolution, blur, gaussian_blur, median, bilateral]

for i in range(6):
    plt.subplot(2,3,i+1), plt.imshow(images[i], "gray")
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()

```
